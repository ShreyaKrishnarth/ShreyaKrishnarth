```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
data = pd.read_csv("snsdata.csv")
data.head()
```

# Treating missing values

```python
data.isnull().sum()
```

```python
data['gender'].value_counts()
```

```python
data['gender'].value_counts(dropna = False)
```

```python
data['gender'].fillna('not disclosed', inplace = True)
```

```python
data['gender'].isnull().sum()
```

```python
data.groupby('gradyear')['age'].mean()
```

```python
data['age'] = data.groupby('gradyear').transform(lambda x : x.fillna(x.mean()))
```

```python
data['age'].isnull().sum()
```

```python
data.isnull().sum()
```

# Treating Outliers

```python
sns.boxplot(data['age'])
```

```python
q1 = data['age'].quantile(0.25)
q3 = data['age'].quantile(0.75)
iqr = q3-q1
```

```python
print(iqr)
```

```python
df = data[(data['age']>(q1 - 1.5*iqr)) & (data['age'] < (q3 + 1.5*iqr))]
```

```python
df['age'].describe()
```

```python
df.shape
```

```python
sns.boxplot(df['age'])
```

# Data Preprocessing

```python
names = df.columns[4:40]
scaled_feature = df.copy()
names
```

```python
features = scaled_feature[names]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(features.values)
```

```python
features = scaler.transform(features.values)
```

```python
scaled_feature[names] = features
scaled_feature.head()
```

# Convert object variables to numeric

```python
def gender_to_numeric(x):
    if x=='M':
        return 1
    if x=='F':
        return 2
    if x=='not disclosed':
        return 3
```

```python
scaled_feature['gender'] = scaled_feature['gender'].apply(gender_to_numeric)
scaled_feature['gender'].head()
```

# Checking the transformed values

```python
scaled_feature.head()
```

# Building the K-means model

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0, n_jobs=-1)
```

```python
model = kmeans.fit(scaled_feature)
```

# Elbow Method

```python
# Creating a function with KMeans to plot "Elbow Curve"
wcss=[]
for i in range(1,20):
    kmeans=KMeans (n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(scaled_feature)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,20),wcss)
plt.title('The Elbow Curve')
plt.xlabel('Number od clusters')
plt.ylabel('WCSS')
plt.show()
```

# Fit K-Means clustering for k=5

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(scaled_feature)
```

```python
len(kmeans.labels_)
```

```python
df['cluster'] = kmeans.labels_
```

```python
df.head()
```
